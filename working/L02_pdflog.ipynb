# PDF of the log

1. Use `numpy` to draw $N$ (where $N$ is large...) random samples from a uniform distribution between $0.1$ and $10$, and store these samples as $x$.
2. Use matplotlib to make a histogram of these samples.
3. Compute the base-10 log of your array $x$, and store this as $y$.
4. Make another histogram for $y$. Using the equation to transform probability distributions, write what the theoretical pdf of $y$ is, and overplot it onto your histogram.
5. Compute the log of the mean of $x$ and the mean of $y$. Now compute the log of the median of $x$ and the median of $y$. 

```python
import numpy as np
import matplotlib.pylab as plt
from numpy import random as rnd

plt.rcParams['figure.figsize'] = [8, 8]
plt.rcParams['font.size'] = 15
```

Creating an histogram for the samples (x) from a uniform distribution between 0.1 and 10.

```python
#samples
x = rnd.uniform(0.1, 10, 10000)

#plot of the histogram
plt.hist(x, bins=100, density=True, histtype='step')
plt.xlabel('x samples')
plt.ylabel('counts')
plt.show()
```

Creating the histogram for the log10 array of samples, called y.

```python
#log10 of the array sample x
y = np.log10(x)

#plot of the new histogram
plt.hist(y, bins=30, density=True, histtype='step')
plt.xlabel('y')
plt.ylabel('counts')
plt.show()
```

Creation of the pdf for the y samples.

```python
#PDF of y
y_grid = np.linspace(-1,1,1000)
pdf_y = (10**(y_grid)*np.log(10))/(10-0.1)

plt.plot(y_grid, pdf_y) 
plt.show()
```

Plotting the pdf of y over its histogram.

```python
plt.hist(y, bins=30, density=True, histtype='step', label='Histogram of y')
plt.plot(y_grid, pdf_y, c='r', label='pdf')
plt.xlabel('y')
plt.ylabel('counts')
plt.legend(loc="upper left")
plt.show()
```

Now I compute the log mean and median for x and y.

```python
#Mean
mean_x = np.log10(np.mean(x))
print("The mean of x is:", mean_x)
mean_y = np.mean(y)
print("The mean of y is:", mean_y)

#Median
median_x = np.log10(np.median(x))
print("The median of x is:", median_x)
median_y = np.median(y)
print("The median of y is:", median_y)
```
