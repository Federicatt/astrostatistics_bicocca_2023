1) Read in `solutions/transient.npy`. This is a (100 x 3) numpy array, containing some time-series measurements from a transient phenomenon. The first column is time (arbitrary units), the second column is the flux (arbitrary units), and the third column are homoescedastic measurement uncertainties on the flux. Plot this data with the uncertainties.

2) One possible model for this data is a sharp burst with an exponential tail rising above a nominal background level. This model can be described as:

$y = b$, for $t<t_0$

$y = b + A\exp[-\alpha(t-t_0)]$, for $t\geq t_0$

where $b$ is the background amplitude, $A$ is the burst amplitude, $t_0$ is the burst epoch, and $\alpha$ encodes how rapidly the burst dies off.

Code this burst model, and code a likelihood function for this data with the burst model. *Hint: This is similar in spirit to the likelihood when fitting polynomials to data, except now $x$ is time, and the polynomial model is now the burst model.* 

3) Using one of the MCMC sampler that we looked at in class (`emcee`, `pymc3`), perform a Bayesian analysis that recovers an MCMC chain of posterior probability samples for the joint parameter space of $\{b,A,t_0,\alpha\}$. You must decide on how many samples are appropriate for running your MCMC. For your priors, make them $b\in \mathrm{Uniform}[0,50]$, $A\in \mathrm{Uniform}[0,50]$, $t_0\in \mathrm{Uniform}[0,100]$, and $\ln\alpha\in \mathrm{Uniform}[-5,5]$.

4) Perform an appropriate check to burn-in your chain, and thin the chain by selecting only points every auto-correlation length. If you are left with a small number of samples after thinning then you may need to run your MCMC for longer in (3).

5) Make a corner plot of the $4$-D parameter space, with labels, $[68,95]\%$ credible levels and quantiles, and titles.

6) Now choose $100$ samples randomly from your burned, thinned chain and plot the $100$ corresponding burst time-series on a figure along with the original data. You may want to choose a transparency for these $100$ burst curves that allows you to see everything nicely. These are showing the posterior spread of model solutions that are consistent with the data.

7) For each parameter, find the median and 90% credible region of the marginalized distribution and quote results as, for instance, $t_0=X^{+Y}_{-Z}$ 

Download the data and plot them with their error.

```python
data = np.load("transient.npy")
#print(data)

t = data[:,0] #time
f = data[:,1] #flux
f_err = data[:,2] #flux error

plt.errorbar(t, f, yerr=f_err, fmt='o', color='b')
plt.xlabel('Time')
plt.ylabel('Flux')
plt.title('Data')
plt.show()
```

We computed the model as a sharp burst with an exponential tail:

$y = b$, for $t<t_0$

$y = b + A\exp[-\alpha(t-t_0)]$, for $t\geq t_0$

Where $b$ is the background amplitude, $A$ is the burst amplitude, $t_0$ is the burst epoch, and $\alpha$ encodes how rapidly the burst dies off.

Then we coded the likelihood of the data. For the priors we made them $b\in \mathrm{Uniform}[0,50]$, $A\in \mathrm{Uniform}[0,50]$, $t_0\in \mathrm{Uniform}[0,100]$, and $\ln\alpha\in \mathrm{Uniform}[-5,5]$.

```python
#burst model
def model(p, t):
    b, A, t_0, alpha = p
    return np.where(t < t_0, b, b+A*np.exp(-alpha*(t-t_0)))

#I implement an initial model with given values
b = 10
A = 5
t_0 = 50
alpha = 0.1
#b, A, t_0, alpha = 10, 5, 50, 0.1
param = [b,  A, t_0, alpha]

flux_model = model(param, t)

plt.plot(t, flux_model, label='Initial Model', color='r')
plt.errorbar(t, f, yerr=f_err, fmt='o', color='b', label='Data')
plt.xlabel('Time')
plt.ylabel('Flux')
plt.legend()
plt.title('Data with initial model')
plt.show()
```

```python
#logarithmic likelihood for the data with the burst model 
def loglikelihood(p, t, f, f_err):
    b, A, t_0, alpha = p
    return -0.5*np.sum((-(f-model(p, t)))**2/f_err**2) #add -0.5*

#logarithmic of the prior
def logprior(p):
    b, A, t_0, alpha = p
    if 0<b<50 and 0<t_0<100 and 0<A<50 and np.exp(-5)<alpha<np.exp(5):
        return -np.log(alpha)
        #return (1/alpha)
    return -np.inf

#logarithmich of the posterior
def mylogPosterior(p, t, f, f_err):
    b, A, t_0, alpha = p
    return loglikelihood(p, t, f, f_err)+logprior(p)
```

# MCMC EMCEE

Using the MCMC sampler emcee I perform a Bayesian analysis that recovers an MCMC chain of posterior probability samples for the joint parameter space of $\{b,A,t_0,\alpha\}$. 

```python
ndim = 4  # number of parameters in the model
nwalkers = 20  # number of MCMC walkers
nsteps = 25000  # number of MCMC steps to take **for each walker**

# initialize theta 
np.random.seed(0)
starting_guesses = param+1e-3*np.random.random((nwalkers, ndim))
```

```python
sampler = emcee.EnsembleSampler(nwalkers, ndim, mylogPosterior, args=[t, f, f_err])
sampler.run_mcmc(starting_guesses, nsteps, progress=True)

print("done")
```

Check of MCMC emcee

```python
fig, axes = plt.subplots(4, figsize=(10, 7), sharex=True)

samples = sampler.get_chain()
print(len(samples))
labels = ["b", "A", "t0", "alpha"]
for i in range(ndim):
    ax = axes[i]
    ax.plot(samples[:, :, i], "k", alpha=0.3)
    ax.set_xlim(0, len(samples))
    ax.set_ylabel(labels[i])
    ax.yaxis.set_label_coords(-0.1, 0.5)

axes[-1].set_xlabel("step number");
```

The MCMC emcee works.
Now I perform a check to burn-in the chain, and thin the chain by selecing points every autocorrelation lenghts.

```python
#autocorrelation length for each parameter
tau = sampler.get_autocorr_time()
print(tau)
```

```python
#burn-in
burnin = 2000

fig, axes = plt.subplots(4, figsize=(10, 7), sharex=True)
samples = sampler.get_chain()
print(len(samples))
labels = ["b", "A", "t0", "alpha"]
for i in range(ndim):
    ax = axes[i]
    ax.plot(samples[:, :, i], "k", alpha=0.3)
    ax.set_xlim(0, len(samples))
    ax.set_ylabel(labels[i])
    ax.axvline(burnin, color='red', alpha=0.5, label='burn-in')
    ax.yaxis.set_label_coords(-0.1, 0.5)

plt.xlim(0, 5000)
axes[-1].set_xlabel("step number");
plt.show()
```

```python
flat_samples = sampler.get_chain(discard=3*int(max(tau)), thin=int(max(tau)), flat=True)
print(flat_samples.shape)
```

# Corner Plot

Now I make a corner plot of the $4$-D parameter space, with labels, $[68,95]\%$ credible levels and quantiles, and titles.

```python
fig = corner.corner(flat_samples, labels=labels, levels=[0.68,0.95], quantiles=[0.16, 0.84], show_titles=True);
```

# Posterior Spread

I choose 100 random samples from the burned, thinned chain and plot the corresponding 100 brust time series on a figure along the original data.

```python
random_samples = flat_samples[np.random.choice(range(len(flat_samples)), 100)]
#print(random_samples)

x_grid = np.linspace(0, 100, 100)

for new_tetha in random_samples:
    y_grid = model(new_tetha, x_grid)
    plt.plot(x_grid, y_grid, alpha=0.2, color='grey')
    
plt.errorbar(t, f, yerr=f_err, fmt='o', color='b', label='Data')
plt.xlabel('Time')
plt.ylabel('Flux')
plt.legend()
plt.title('Posterior Spread')
plt.show()
```

The model seems to be consistent with the data. Now for each parameter I find the median and 90% credible region of the marginalized distribution and quote results as $t_0=X^{+Y}_{-Z}$ 

```python
median = np.empty(ndim)
quantile_95 = np.empty(ndim) 
quantile_05 = np.empty(ndim) 

for i in range(ndim):
    median[i] = np.median(flat_samples[:, i])
    quantile_95[i] = np.quantile(flat_samples[:, i], 0.95)
    quantile_05[i] = np.quantile(flat_samples[:, i], 0.05)
    
    up = quantile_95[i] - median[i]
    low = median[i] - quantile_05[i]
    
    print(labels[i],'=', str(median[i]),'+', str(up), '-', str(low))
    
```
