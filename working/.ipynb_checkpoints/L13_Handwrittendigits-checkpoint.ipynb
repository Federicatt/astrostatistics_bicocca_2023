# Handwritten Digits

Here we're going to use the [MNIST database](https://en.wikipedia.org/wiki/MNIST_database). This contains a set of images with handwritten digts. It is one of the most widely used and best studied datasets in the whole of machine learning. It's part of the standard tests new tecniques need to face. 

Let's load the dataset first.

```python
from sklearn import datasets
from sklearn.manifold import Isomap
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix

import numpy as np
import matplotlib.pyplot as plt

plt.rcParams['figure.figsize'] = [8, 8]
plt.rcParams['font.size'] = 15
```

```python
digits = datasets.load_digits()
print(digits.images.shape)
print(digits.keys())
```

```python
print(digits.data.shape)
print(digits.target.shape)
```

```python
x = digits.data
print("The data:",x)


y = digits.target
print('The targets:', y)
```

# Part 1

We'd like to visualize our points within the 64-dimensional parameter space, but it's difficult to plot points in $N=64$ dimensions! So we need to reduce the dimensionality of the space.


3) Use the unsupervised dimensionality reduction technique called **[`Isomap`](http://scikit-learn.org/stable/modules/generated/sklearn.manifold.Isomap.html#sklearn.manifold.Isomap)**. Isomap is an embedding tecnique, which tries to understand if your $N$ dimensional datapoint live in a $N$ dimensional (with $N<M$) manyfold. Have a look at the Isomap documentation and reduce the MINST dataset.  Do some plots to visualize your results. 

```python
#Reduction with Isomap

isomap = Isomap(n_components=2)
x_transformed = isomap.fit_transform(x)
x_transformed.shape
```

```python
#Plot the reduced data

plt.scatter(x_transformed[:, 0], x_transformed[:, 1], color='b', alpha=0.3)
plt.title('Reduced Data')
plt.show()
```

```python
#Plot the reduced data with coloring the poits by their labels

plt.scatter(x_transformed[:,0], x_transformed[:,1], c=y, edgecolor='none', alpha=0.7, cmap=plt.cm.get_cmap('Paired', 10))
plt.colorbar(label='digit label', ticks=range(10))
plt.title('Reduced Data')
plt.show()
```

# Part 2

4) Split the dataset taking 80% of the images as tranining dataset and leaving the other 20% as a validation set. You need sklearn.model_selection.train_test_split.

I start by splitting the dataset with sklearn.model_selection.train_test_split and taking the 80% of the images as training dataset.

```python
xtrain, xtest, ytrain, ytest = train_test_split(digits.data, digits.target, random_state=2, train_size=0.8)
print(xtrain.shape)
print(xtest.shape)
```

5) Use a simple Logistic Regression which despite its name is a common classification algorithm. (I know, this is confusing!). Use the flag solver='sag', the default solver had a bug, at least in my version.

```python
#Logistic Regression

LogRegr = LogisticRegression(max_iter=2000, solver='sag')
LogRegr.fit(xtrain, ytrain)

#predict the labels
train_predicted = LogRegr.predict(xtrain)
test_predicted = LogRegr.predict(xtest)
```

6) Compute the accuracy (sklearn.metrics.accuracy_score) of your emulator. In this context, accuracy is simply the fraction of the digits that were correctly identified. Compute the accuracy on both the training and the validation set. Comment on the results.

```python
#accuracy test
accuracy_train = accuracy_score(ytrain, train_predicted)
accuracy_test = accuracy_score(ytest, test_predicted)

print('The accuracy of the train test is:', accuracy_train)
print('The accuracy of the test set is:', accuracy_test, '\n')
```

From the accuracy test we can see that we have almost 95% of the digits classified correctly.

7) The accuracy is a single number doesn't tell us where we've gone wrong. Plot the confusion matrix as we did above. What are the digits that are more challenging to identify, does it make sense?

Through the confusion matrix we can see which are the numbers that were misclassified.

```python
#confusion matrix
confusion_train = confusion_matrix(ytrain, train_predicted)
confusion_test = confusion_matrix(ytest, test_predicted)

print('The confusion matrix for the train test is: \n', confusion_train, '\n')
print('The confusion matrix for the test set is: \n', confusion_test)
```

In this case the numbers that were miscalssified are: 
- one 0 classified as a 4 (out of 31 zeros classified correctly);
- three 1 classified as a 3, as a 8 and as a 9;
- two 5 classified as 1 and as 4;
And so on.
